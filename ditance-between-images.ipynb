```python

#function returns the distance between two images giving their index i and j
def distance(i,j):
    element1=train_data[i]
    element2=train_data[j]
    return np.sum(np.square(element1-element2))

#x=distance(5,150)
#print(x)

#we define another function in order to find the distance between two elements since
#we are working with vectors(each line is an image)
def distance(x,y):
#function that returns the distance between two vectors passed in arguments
    return np.sum(np.square(x,y))

#Nearest neighbour function working with the previously defined distance function
def NN(x):
    tab=[]
    for i in range (len(train_data)):
        tab.append(distance(train_data[i],x))
    return np.argmin(tab)

NN(train_data[50])
```
